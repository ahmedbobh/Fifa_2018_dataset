# Data Mining Project with Hierarichal and K-medoids Clustering

## we import our libraries

```python
import numpy as np
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
wc_data= pd.read_csv('all_wc_18_players_fifa.csv')
```

```python
wc_data.shape
```

```python
wc_data.head
```

```python
data = wc_data.iloc[:,6:8].values
```

```python
dendrogram=sch.dendrogram(sch.linkage(data[:15],'single'))
plt.title('dendrogram')
plt.xlabel('X values')
plt.ylabel('Distance')
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean',linkage='single')
cluster.fit_predict(data)
```

```python
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
plt.title('Clusters of players')
```
